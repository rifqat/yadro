```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# Ma'lumotlarni graf ko'rinishida yaratish
def create_graph_with_weights(data, epsilon):
    G = nx.Graph()  # Grafni boshlash
    distance_matrix = pdist(data, metric='euclidean')
    A = squareform(distance_matrix)
    max_vals = np.max(A, axis=1, keepdims=True)
    weight = (max_vals - A) / max_vals
    #np.fill_diagonal(weight, 0)
    
    # Qirralarni vaznlari bilan yaratish
    for i in range(len(data)):
        for j in range(i + 1, len(data)):
            #distance = A[i, j]  # Nuqtalar orasidagi masofa
            # w_ij ni hisoblash
            #weight = (max_distance - distance) / max_distance
            if weight[i,j] > epsilon:
                G.add_edge(i, j, weight=weight[i,j])  # Qirrani vazni bilan qo'shish
            
    return G
```

```python
import heapq  # Fibonacci heap uchun yaqin o'xshashlik

# Zichlikni hisoblash
def compute_local_density(G):
    density = {}
    for node in G.nodes:
        # Tugunning qo'shnilari bilan qirra vaznlarining yig'indisi zichlik sifatida olinadi
        density[node] = sum(G[node][neighbor]['weight'] for neighbor in G.neighbors(node))
    return density

# Zichlik o'zgarishlar ketma-ketligini hisoblash
def compute_density_variation_sequence(G):
    # Fibonacci heapga o'xshash heapq kutubxonasidan foydalanamiz
    density = compute_local_density(G)
    H = list(G.nodes)
    heap = [(density[node], node) for node in H]
    heapq.heapify(heap)  # Minimal zichlikni topish uchun heapni yaratamiz

    Dt_sequence = []  # Zichlik o'zgarishlari ketma-ketligi
    Mt_sequence = []  # Eng zaif tugunlar ketma-ketligi

    while heap:
        Dt, Mt = heapq.heappop(heap)  # Eng kichik zichlikli tugunni olish
        Dt_sequence.append(Dt)  # Zichlik qiymatini saqlash
        Mt_sequence.append(Mt)  # Tugunni saqlash

        # Qo'shni tugunlarni olishdan oldin ularning zichligini yangilaymiz
        neighbors = list(G.neighbors(Mt))

        for neighbor in neighbors:
            if neighbor in density:
                density[neighbor] -= G[Mt][neighbor]['weight']

        # Tugunni olib tashlaymiz
        G.remove_node(Mt)

        # Yangilangan zichliklarni heapga qayta joylaymiz
        heap = [(density[node], node) for node in G.nodes]
        heapq.heapify(heap)

    return Dt_sequence, Mt_sequence

# Misol tasvir (10x10 grayscale)
#image = np.random.rand(10, 10)  # Random 10x10 tasvir

# Proximity grafini qurish
#G = build_proximity_graph(image, sigma_I, sigma_d, r)

# Zichlik o'zgarishlar ketma-ketligini hisoblash
#Gr=G.copy()

#Dt_sequence, Mt_sequence = compute_density_variation_sequence(Gr)


# Natijalarni chop etish
#print("Density variation sequence (Dt):", Dt_sequence)
#print("Weakest pixel sequence (Mt):", Mt_sequence)
#print(len(Mt_sequence))
```

```python
# Asosiy piksellarni aniqlash funksiyasi
def identify_core_pixels(Dt_sequence, Mt_sequence, delta, beta):
    # Pasayish tezligini hisoblash: R_t = (D_t - D_(t+1)) / D_t
    Rt_sequence = [(Dt_sequence[t] - Dt_sequence[t+1]) / Dt_sequence[t] for t in range(len(Dt_sequence)-1)]
    
    # Ijobiy R_t qiymatlarini saralash
    positive_Rt = [r for r in Rt_sequence if r > 0]
    positive_Rt_sorted = sorted(positive_Rt)

    # R_t ning δ pozitsiyasiga mos keladigan α ni aniqlash
    alpha_index = int(len(positive_Rt_sorted) * delta)
    alpha = positive_Rt_sorted[alpha_index]

    # Asosiy piksellar to'plamini aniqlash
    core_pixels = []
    consecutive_count = 0

    for t in range(len(Rt_sequence)):
        if Rt_sequence[t] > alpha:
            consecutive_count += 1
            if consecutive_count >= beta:
                core_pixels.append(Mt_sequence[t])
        else:
            consecutive_count = 0  # Ketma-ketlik tugadi, yana boshdan boshlaymiz

    return core_pixels

# Misol uchun δ va β parametrlari
#delta = 0.5  # Pasayish darajasi
#beta = 5   # Ketma-ket kamida 3 tugun bo'lishi kerak

# Zichlik o'zgarishlar ketma-ketligini hisoblash (Dt_sequence va Mt_sequence)
#Dt_sequence = [...]  # Avvalgi hisoblangan zichliklar ketma-ketligi
#Mt_sequence = [...]  # Avvalgi eng zaif tugunlar ketma-ketligi

# Asosiy piksellarni aniqlash
#core_pixels = identify_core_pixels(Dt_sequence, Mt_sequence, delta, beta)

# Natijalarni chop etish
#print("Core Pixels:", core_pixels)
```

```python
import networkx as nx

# Asosiy piksellarning grafini segmentlarga bo'lish funksiyasi
def partition_core_pixels(G, core_pixels, theta=0.1):
    # Core piksellardan hosil bo'lgan subgrafni yaratish
    core_graph = G.subgraph(core_pixels).copy()

    # Zaif qirralarni olib tashlash (vazni theta dan kichik bo'lgan qirralar)
    weak_edges = [(u, v) for u, v, w in core_graph.edges(data='weight') if w < theta]
    core_graph.remove_edges_from(weak_edges)

    # Grafning ulanmagan komponentlarini topish
    segments = list(nx.connected_components(core_graph))

    return segments

# Asosiy piksellarni segmentlash uchun funktsiyani chaqirish
# Core_pixels bu oldin aniqlangan asosiy piksellar to'plami
# G bu proximity graph (yaqinlik grafigi)

#core_segments = partition_core_pixels(G, core_pixels, theta=0)

# Natijalarni chop etish
#print("Number of segments found:", len(core_segments))
#for i, segment in enumerate(core_segments):
#    print(f"Segment {i+1}: {segment}")
```

```python
import numpy as np

# Funksiya piksel va segment o'rtasidagi o'xshashlikni hisoblaydi
def calculate_similarity(G, pixel, segment):
    weights = [G[pixel][s]['weight'] for s in segment if G.has_edge(pixel, s)]
    if weights:
        return np.mean(weights)
    return 0

# Tasvir segmentatsiyasi uchun kengaytirish funksiyasi
def expand_segments(G, Mt_sequence, core_segments, lambda_value=0.5):
    n = len(Mt_sequence)
    segments = [list(segment) for segment in core_segments]  # Core segmentlarni ro'yxat sifatida boshlaymiz
    low_confidence_pixels = set()  # Past ishonchli piksellar to'plami
    added_pixels = set()  # Qo'shilgan tugunlarni kuzatib borish uchun to'plam

    # Orqaga ketma-ketlikda ishlaymiz
    for t in range(n-1, -1, -1):
        Mt = Mt_sequence[t]

        # Agar Mt hali segmentga qo'shilmagan bo'lsa
        if Mt not in added_pixels:
            # Har bir segment bilan o'xshashlikni hisoblaymiz
            similarities = [(s, calculate_similarity(G, Mt, seg)) for s, seg in enumerate(segments)]
            similarities = [(s, sim) for s, sim in similarities if sim > 0]

            if similarities:
                # Eng o'xshash segmentni topamiz
                similarities.sort(key=lambda x: x[1], reverse=True)
                m1 = similarities[0][1]
                s1 = similarities[0][0]

                # Ikkinchi o'xshash segmentni topamiz
                m2 = similarities[1][1] if len(similarities) > 1 else 0

                # Pikselni eng o'xshash segmentga qo'shamiz
                if m1 > 0:
                    segments[s1].append(Mt)  # Segmentga yangi piksel qo'shamiz
                    added_pixels.add(Mt)  # Qo'shilgan piksellarni kuzatamiz

                    # Agar ikkinchi o'xshash segment yaqin bo'lsa, past ishonchli piksellarga qo'shamiz
                    if m2 > lambda_value * m1:
                        low_confidence_pixels.add(Mt)
            else:
                # Yangi segment yaratamiz, chunki u hali segmentga mos kelmadi
                segments.append([Mt])
                added_pixels.add(Mt)  # Qo'shilgan piksellarni kuzatamiz

    # Tugunni faqat bir marta qo'shganimizni tekshiramiz
    for i in range(len(segments)):
        segments[i] = list(set(segments[i]))  # Har bir segmentdan takrorlangan tugunlarni olib tashlaymiz

    # Grafdagi barcha tugunlar qo'shilganini tekshirish
    all_nodes = set(G.nodes)
    added_nodes = set().union(*segments)
    
    # Qo'shilmay qolgan tugunlarni topamiz
    missing_nodes = all_nodes - added_nodes

    # Har qanday qo'shilmagan tugunlarni yangi segment sifatida qo'shamiz
    for node in missing_nodes:
        print(f"Tugun {node} qo'shilmagan, yangi segmentga qo'shilmoqda.")
        segments.append([node])

    return segments, low_confidence_pixels
# Misol uchun ketma-ketlik va core segmentlar
#Mt_sequence = [...]  # Avvalgi hisoblangan zaif tugunlar ketma-ketligi
#core_segments = [...]  # Avvalgi core segmentlar
#print(len(core_segments))
# Segmentatsiyani kengaytirish
#segments, low_confidence_pixels = expand_segments(G, Mt_sequence, core_segments, lambda_value=0.5)
#print(len(segments))
# Natijalarni chop etish
#print("Final segments:", segments)
#print("Low confidence pixels:", low_confidence_pixels)

# Barcha tugunlar qo'shilganini tekshirish
#all_nodes = set(G.nodes)
#added_nodes = set().union(*segments)
#if all_nodes == added_nodes:
#    print("Barcha tugunlar segmentlarga kiritildi.")
#else:
#    print("Ba'zi tugunlar segmentlarga kiritilmadi.")
```

```python
import networkx as nx
import numpy as np

def compute_DVI(G, clusters):
    # Har bir tugunni o'z klasteriga xaritalash
    node_to_cluster = {}
    for i, cluster in enumerate(clusters):
        for node in cluster:
            node_to_cluster[node] = i

    # Intra va inter zichliklarini boshlang'ichlash
    intra_density = {node: 0 for node in G.nodes()}
    inter_density = {node: 0 for node in G.nodes()}

    # Qirralarni ko'rib chiqish va vaznlarni yig'ish
    for u, v, data in G.edges(data=True):
        weight = data.get('weight', 1)
        if node_to_cluster[u] == node_to_cluster[v]:
            intra_density[u] += weight
            intra_density[v] += weight
        else:
            inter_density[u] += weight
            inter_density[v] += weight

    # DVI_1 ni hisoblash
    dvi1_value = sum(intra_density.values()) - sum(inter_density.values())

    # DVI_2 ni hisoblash
    intra_sum = sum(intra_density.values())
    inter_sum = sum(inter_density.values())
    dvi2_value = inter_sum / intra_sum if intra_sum > 0 else float('inf')

    # DVI_3 ni hisoblash
    dvi3_value = 0
    for node in G.nodes():
        if intra_density[node] > 0:
            dvi3_value += inter_density[node] / intra_density[node]
    
    return dvi1_value, dvi2_value, dvi3_value

# # DVI metrikalarini hisoblash
# dvi1_value, dvi2_value, dvi3_value, tightness_value = compute_DVI_and_tightness(G, clusters)

# # Natijalarni chop etish
# print(f"DVI_1: {dvi1_value}")
# print(f"DVI_2: {dvi2_value}")
# print(f"DVI_3: {dvi3_value}")
```

```python


def visualize_clusters(data,G,segments):
    # Klasterlarni tasvirlash    
    pos = {i: data[i] for i in range(len(data))}
    colors = ['r', 'g', 'b', 'y', 'c', 'm']
    for idx, cluster in enumerate(segments):
        nx.draw_networkx_nodes(G, pos, nodelist=cluster, node_color=colors[idx % len(colors)], node_size=25)
    #nx.draw_networkx_nodes(G, pos, nodelist=low_confidence_pixels, node_color=colors[5], node_size=25)
    nx.draw_networkx_edges(G, pos, alpha=0.25)
    plt.show()
```

```python


def yadro(data,Gr,Dt_sequence, Mt_sequence,delta,beta,visualize = True):
   
    # Asosiy piksellarni aniqlash
    core_pixels = identify_core_pixels(Dt_sequence, Mt_sequence, delta, beta)
    core_segments = partition_core_pixels(Gr, core_pixels, theta=0.1)
    segments, low_confidence_pixels = expand_segments(Gr, Mt_sequence, core_segments, lambda_value=0.5)
    dvi1, dvi2, dvi3 = compute_DVI(Gr, segments)
    
    if visualize:
        visualize_clusters(data,Gr, segments)
    
    return segments, dvi1, dvi2, dvi3
```

```python
# Function to find optimal clustering parameters
def find_optimal_parameters(data,delta_values,beta):
    best_params = None
    best_score = np.inf
    epsilon = 0.8
    # Vaznlangan graf yaratish
    G = create_graph_with_weights(data,epsilon)
    Gr=G.copy()
    Dt_sequence, Mt_sequence = compute_density_variation_sequence(Gr)
    # Misol uchun δ va β parametrlari
    #delta = 0.8  # Pasayish darajasi
    #beta = 5   # Ketma-ket kamida 3 tugun bo'lishi kerak

    for delta in delta_values:
        clusters, dvi1, dvi2, dvi3 = yadro(data,G,Dt_sequence, Mt_sequence,delta,beta)
         # Normalize and combine DVI scores and tightness_avg to get a single score
        
        if dvi3 < best_score:
            best_score = dvi3
            best_params = delta
        print("alfa={0} va best_score={1}".format(best_params, best_score))

    return best_params, best_score
```

```python
data,y=datasets[5][0]
alfa = np.arange(0, 1, 0.1)
beta=5
best_params, best_score = find_optimal_parameters(data,alfa,beta)
print("Best Parameters:", best_params)
print("Best Score:", best_score)
```
